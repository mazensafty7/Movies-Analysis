# Data Cleaning

### Correct Format

```python
movies["title_year"].info()
```

```python
movies["title_year"] = pd.to_datetime(movies["title_year"] , format = "%Y")
movies["title_year"].info()
movies.head(5)
```

### Check For Duplicates 

```python
movies.duplicated().sum()
```

### Hundle Missing Vlaues

```python
print(movies.isnull().sum() , "\n")

percentage = movies.isnull().sum() / movies.shape[0] * 100
print(percentage)
```

#### aspect_ratio is normaly distribuated so the best way to fill the column with the mean of that column

```python
movies["aspect_ratio"] = movies["aspect_ratio"].fillna(movies["aspect_ratio"].mean())

print(movies.isnull().sum() , "\n")
```

### Title Year Column

```python
movies["title_year"].mode()
```

```python
movies["title_year"] = movies["title_year"].fillna(movies["title_year"].mode()[0])
```

### Content Rating Column

```python
movies["content_rating"].mode()
movies["content_rating"].value_counts()
```

```python
movies["content_rating"] = movies["content_rating"].fillna(movies["content_rating"].mode()[0])
```

### Duration Year Column

```python
movies["duration"] = movies["duration"].fillna(movies["duration"].mean())
```

### Color Column

```python
movies["color"].mode()
```

```python
movies["color"] = movies["color"].fillna(movies["color"].mode()[0])
```

###  facenumber_in_poster Column

```python
movies["facenumber_in_poster"] = movies["facenumber_in_poster"].fillna(movies["facenumber_in_poster"].mean())
```

###  Country Column

```python
movies["country"].mode()
```

```python
movies["country"] = movies["country"].fillna(movies["country"].mode()[0])
print(movies.isnull().sum())
```

###  Language Column

```python
movies["language"].mode()
```

```python
movies["language"] = movies["language"].fillna(movies["language"].mode()[0])
```

### num_user_for_reviews column

```python
movies["num_user_for_reviews"] = movies["num_user_for_reviews"].fillna(movies["num_user_for_reviews"].mean())
```

### plot_keywords column

```python
movies["plot_keywords"].mode()
```

```python
movies["plot_keywords"] = movies["plot_keywords"].fillna(movies["plot_keywords"].mode()[0])
```

```python
print(movies.isnull().sum())
```

### gross and budget column

```python
movies["gross"] = movies["gross"].fillna(movies["gross"].mean())
movies["budget"] = movies["budget"].fillna(movies["budget"].mean())
print(movies.isnull().sum())
```

### Drop Null rows

```python
movies.dropna(inplace = True)
print(movies.isnull().sum())
```

## Director Name

```python
movies["director_name"].value_counts()
```

```python
plt.figure(figsize = [25 , 25])

plt.subplot(2 , 2 , 1)
direc_gross = movies.groupby(["director_name"])["gross"].sum()
top_3_directors = direc_gross.nlargest(10)
top_3_directors = top_3_directors.reset_index()
sb.barplot(data = top_3_directors , x = "director_name" , y = "gross")
plt.xticks(rotation = 90)
plt.title("Top 10 Directors Gross")

plt.subplot(2 , 2 , 2)
smallest_3_directors = direc_gross.nsmallest(10)
smallest_3_directors = smallest_3_directors.reset_index()
sb.barplot(data = smallest_3_directors , x = "director_name" , y = "gross")
plt.xticks(rotation = 90)
plt.title("Smallest 10 Directors Gross")

plt.subplot(2 , 2 , 3)
direc_budget = movies.groupby(["director_name"])["budget"].sum()
top_3_directorsb = direc_budget.nlargest(10)
top_3_directorsb = top_3_directorsb.reset_index()
sb.barplot(data = top_3_directorsb , x = "director_name" , y = "budget")
plt.xticks(rotation = 90)
plt.title("Top 10 Directors Budget")

plt.subplot(2 , 2 , 4)
smallest_3_directorsb = direc_budget.nsmallest(10)
smallest_3_directorsb = smallest_3_directorsb.reset_index()
sb.barplot(data = smallest_3_directorsb , x = "director_name" , y = "budget")
plt.xticks(rotation = 90)
plt.title("Smallest 10 Directors Budget");
```
